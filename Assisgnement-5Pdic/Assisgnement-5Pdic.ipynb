### 1) Repeat a tuple three times using *

```python
# Original tuple
t = (1, 2, 3)

# Repeat tuple three times
result = t * 3

print("Repeated Tuple:", result)
```

#### 2) Join three separate tuples into one using +

```python
t1 = (1, 2)
t2 = (3, 4)
t3 = (5, 6)

# Join tuples
result = t1 + t2 + t3

print("Joined Tuple:", result)
```

### 3) Check whether a specific element exists in a tuple using in

```python
t = (10, 20, 30, 40)

element = 30

if element in t:
    print(f"{element} exists in the tuple.")
else:
    print(f"{element} does not exist in the tuple.")
```

#### 4) Calculate total, highest, and lowest without sum(), max(), min()

```python
t = (5, 12, 7, 3, 9)

# Initialize values
total = 0
highest = t[0]
lowest = t[0]

for num in t:
    total += num
    if num > highest:
        highest = num
    if num < lowest:
        lowest = num

print("Total:", total)
print("Highest:", highest)
print("Lowest:", lowest)
```

### 5) Filter tuple values greater than 10

```python
n = (3, 14, 7, 22, 9, 41, 18, 5)

# Use tuple comprehension
filtered = tuple(x for x in n if x > 10)

print("Filtered Tuple:", filtered)
```

### 6) Count elements in a set without len()

```python
s = {"cat", "dog", "bird", "fish"}

count = 0
for _ in s:
    count += 1

print("Number of elements in set:", count)
```

### 7) Combine two sets into one (union)

```python
s1 = {1, 2, 3}
s2 = {3, 4, 5}

combined = s1 | s2   # union operator

print("Combined Set:", combined)
```

### 8) Find common elements in two sets (intersection)

```python
s1 = {1, 2, 3, 4}
s2 = {3, 4, 5, 6}

common = s1 & s2   # intersection operator

print("Common Elements:", common)
```

#### 9) Find all elements that are in either set (union)

```python
s1 = {1, 2, 3, 4}
s2 = {3, 4, 5, 6}

either = s1 | s2   # union operator

print("Elements in either set:", either)
```
